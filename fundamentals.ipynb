### 1. Create a random tensor with shape `(7, 7)`.


````python
# Import torch
import torch

# Create random tensor
t = torch.rand(size = (7,7))
````

### 2. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 7)`

````python
# Create another random tensor
t2 = torch.rand(size = (1,7))

# Perform matrix multiplication
torch.matmul(t, t2.T)
````

### 3. Set the random seed to `0` and do 1 & 2 over again.

The output should be:
```
(tensor([[1.8542],
         [1.9611],
         [2.2884],
         [3.0481],
         [1.7067],
         [2.5290],
         [1.7989]]), torch.Size([7, 1]))
```

````python
torch.manual_seed(0)

# Create two random tensors
t3 = torch.rand(size = (7,7))
t4 = torch.rand(size = (1,7))
# Matrix multiply tensors
torch.matmul(t3, t4.T)
````


### 5. Workng with GPU

````python
# Set random seed
torch.manual_seed(1234)

# Check for access to GPU
torch.cuda.is_available()
# Create two random tensors on GPU
t5 = torch.rand(size = (2,3), device = 'cuda')
t6 = torch.rand(size = (2,3), device = 'cuda')
````


### 6. Perform a matrix multiplication on the GPU

````python
# Perform matmul on tensor_A and tensor_B
result = torch.matmul(t5, t6.T)
result
````

### 7. Find the maximum and minimum values of tensor

````python
# Find max
print(result.max())
# Find min
print(result.min())
````

### 8. Find the maximum and minimum index values

````python
# Find arg max
print(result.argmax())

# Find arg min
print(result.argmin())
````


### 9. Squeezing and unsqueezing

````python
# Set seed
torch.manual_seed(7)

# Create random tensor
t_final = torch.rand(size = (1, 1, 1, 10))

# Remove single dimensions
t_final_removed = torch.squeeze(t_final)

# Print out tensors and their shapes
print(f'Before squeezing: {t_final} of shape {t_final.shape}')
print()
print(f'After squeezing: {t_final_removed} of shape {t_final_removed.shape}')
````
